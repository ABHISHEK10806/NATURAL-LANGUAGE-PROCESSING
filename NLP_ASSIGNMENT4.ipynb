# 6) NLP Preprocessing (Simple Code)

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

# 7) One Sample Review

```python
text = df['review'][0]

print("Original Review:\n", text)
```

# 8) Tokenization

```python
tokens = word_tokenize(text)
print("\nTokens:\n", tokens[:20])
```

# 9) Stopword Removal

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [
    word for word in tokens
    if word.lower() not in stop_words and word.isalpha()
]

print("\nAfter Stopword Removal:\n", filtered_tokens[:20])
```

# 10) Stemming

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_tokens]

print("\nAfter Stemming:\n", stemmed_words[:20])
```

# 11) Lemmatization

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\nAfter Lemmatization:\n", lemmatized_words[:20])
```
